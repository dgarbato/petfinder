```python
# importing the libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
  
# loading the dataset
# from seaborn library
data = sns.load_dataset('car_crashes')
  
# viewing the dataset
print(data.head(4))
```

```python
# plotting the density plot 
# for 'speeding' attribute
# using plot.density()
data.speeding.plot.density(color='green')
plt.title('Density plot for Speeding')
plt.show()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
sns.set_style('white')
```

```python
g = sns.FacetGrid(penguins,col='island'); #you can also use row
g.map(sns.histplot,"flipper_length_mm")  #uses positional arguments
```

```python
g = sns.FacetGrid(penguins,col='island'); #you can also use row
g.map_dataframe(sns.histplot, y='flipper_length_mm');
```

```python
type(g)
```

```python
g = sns.FacetGrid(penguins,col='island') #you can also use row
g.map_dataframe(sns.scatterplot, x='bill_depth_mm', y= 'bill_length_mm')
g.set_axis_labels('Culment Depth (mm)', 'Culmen Length (mm)')
g.set_titles(col_template= '{col_name} Island');
```

```python
g = sns.FacetGrid(penguins,col='island',row='species') #you can also use row
g.map_dataframe(sns.scatterplot, x='bill_depth_mm', y= 'bill_length_mm')
g.set_axis_labels('Culment Depth (mm)', 'Culmen Length (mm)')
g.set_titles(col_template= '{col_name} Island',row_template= '{row_name}');
```

```python
g = sns.FacetGrid(penguins,col='island',sharey=False,ylim=(20,70)) #you can also use row

g.map_dataframe(sns.scatterplot, x='bill_depth_mm', y= 'bill_length_mm')
g.set_axis_labels('Culment Depth (mm)', 'Culmen Length (mm)')
g.set_titles(col_template= '{col_name} Island');
```

```python
g = sns.FacetGrid(penguins,col='island',hue='species',palette='prism') #put hue here
g.map_dataframe(sns.scatterplot, x='bill_depth_mm', y= 'bill_length_mm')
g.set_axis_labels('Culment Depth (mm)', 'Culmen Length (mm)')
g.set_titles(col_template= '{col_name} Island')
g.add_legend();
```

```python
quantity='body_mass_g'

g = sns.FacetGrid(penguins, row='sex', col='species')
g.map_dataframe(sns.kdeplot, x=quantity, lw=3, shade=True)
g.set_titles(row_template='{row_name}', col_template='{col_name} Penguins')
```
