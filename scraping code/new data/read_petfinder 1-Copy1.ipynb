```python
import pandas as pd
import numpy as np
import re
```

```python
def read_pet(file):
    return(
    pd.read_csv(file,
               names=['cat_name', 'num_images', 'num_videos', 'age', 'breed',
                      'color', 'size', 'sex', 'characteristics', 'blank_coat_length',
                      'house_trained', 'health', 'good_in_a_home_with', 'adoption_fee',
                      'prefers_a_home_without', 'coat_length', 'pet_story_num_words',
                      'rescue_group', 'days_on_petfinder', 'special_needs']
               ))
```

```python
def get_data(file,info):
    dsn = read_pet(file)
    dsn['data_set'] = info
    
    mask=dsn['special_needs'].isnull()
    
    dsn.loc[mask,'special_needs'] = dsn.loc[mask,'days_on_petfinder'] 
    dsn.loc[mask,'days_on_petfinder'] = dsn.loc[mask,'rescue_group'] 
    dsn.loc[mask,'rescue_group'] = dsn.loc[mask,'pet_story_num_words'] 
    dsn.loc[mask,'pet_story_num_words'] = dsn.loc[mask,'coat_length']  
    dsn.loc[mask,'coat_length'] = np.nan
    
    dsn['pet_story_num_words'] = pd.to_numeric(dsn['pet_story_num_words'],errors = 'coerce')
    return(dsn)
    
```
